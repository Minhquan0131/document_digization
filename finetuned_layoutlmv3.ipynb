## **Convert COCO json file to masked**

```python
from matplotlib import pyplot as plt
from pycocotools.coco import COCO
from pycocotools import mask
import os
from PIL import Image, ImageDraw, ImageFont, ImageOps
import json
import numpy as np

```

```python
def list_files(in_path):
    img_files = []
    for (dirpath, dirnames, filenames) in os.walk(in_path):
        for file in filenames:
            filename, ext = os.path.splitext(file)
            ext = str.lower(ext)
            if ext == '.jpg' or ext == '.jpeg' or ext == '.gif' or ext == '.png' or ext == '.pgm':
                img_files.append(os.path.join(dirpath, file))
    return img_files
def apply_mask(bin_arr, arr, height, width,category_dict, mask_with_color):
  if mask_with_color:
    canvas = np.zeros((height, width, 4), dtype='uint8')
    for i in range(height):
      for j in range(width):
        if arr[i][j] == 0:
          canvas[i][j] = [255,255,255, 255]
        elif arr[i][j]<=5:
          mlist = list(category_dict[arr[i][j]-1])
          canvas[i][j] = mlist + [255]
    return Image.fromarray(canvas, mode = 'RGBA')


  else:
    return  ImageOps.invert(Image.fromarray(bin_arr*255)).convert("RGBA")
```

```python
json_path = '/content/drive/MyDrive/layoutlmv3-base-finetuned-publaynet/my_test_model/output/coco_instances_results.json'
img_folder_path = '/content/drive/MyDrive/layoutlmv3-base-finetuned-publaynet/my_test_model/input_1910/'
output_folder_path = '/content/drive/MyDrive/layoutlmv3-base-finetuned-publaynet/my_test_model/output_2410_pp/'
font_path = '/content/drive/MyDrive/unilm/layoutlmv3/data/examples/DejaVuSans.ttf'
img_list = list_files(img_folder_path)

CATEGORY2COLOR = {0 : (255, 0, 0),
                  1 : (0, 0, 255),
                  2 : (0, 255, 0),
                  3 : (251, 122, 6),
                  4 : (128, 0, 128)}
CATEGORY2LABEL = {0 : "text",
                  1 : "title",
                  2 : "list",
                  3 : "table",
                  4: "figure"}
HAS_MASK = False
MASK_WITH_COLOR = True
FONT_SIZE = 23
CONFIDENCE_SCORE = 0.7
BBOX_WIDTH = 4


font = ImageFont.truetype(font_path, FONT_SIZE)
json_file = open(json_path)
data = json.load(json_file)

for idx, image_path in enumerate(img_list):
    img = Image.open(image_path).convert('RGBA')
    file_name = os.path.basename(image_path).split('/')[-1]
    segmentation = binary_segmentation = np.zeros([img.height, img.width], dtype='uint8')
    for item in data:

      if item['image_id'] == file_name and item['score']>=CONFIDENCE_SCORE:
        box = item['bbox']
        category = item['category_id']
        score = ": "+str(round(item['score'],3))
        segment = item['segmentation']

        segmentation = segmentation + mask.decode(segment)*(category+1)
        binary_segmentation = binary_segmentation + mask.decode(segment)
        bitmap = Image.fromarray(segmentation)
        draw = ImageDraw.Draw(img)
        draw.rectangle([(box[0],box[1]),(box[2]+box[0],box[3]+box[1])],outline=CATEGORY2COLOR[category], width = BBOX_WIDTH)
        draw.text((box[0]-25,box[1]-25), CATEGORY2LABEL[category] + score, fill=CATEGORY2COLOR[category], font=font)
        # draw.bitmap([(0, 0), (img.width, img.height)], bitmap, fill=CATEGORY2COLOR[category])
    if HAS_MASK:
      map = apply_mask(binary_segmentation,segmentation, img.height, img.width, CATEGORY2COLOR, MASK_WITH_COLOR)
      img = Image.blend(map, img, alpha = 0.8)
    display(img)
    file_name, ext = os.path.splitext(file_name)
    img = img.save(output_folder_path + "mask_"+file_name+".png")

```
